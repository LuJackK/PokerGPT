# PokerGPT model architecture

Instantiate the nanoGPT-style decoder model from preprocessing metadata. Training and evaluation loops will be added after the first processed dataset is available.

```python
from pathlib import Path
import pickle
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT))

from poker_model import GPT, GPTConfig
```

```python
meta_path = PROJECT_ROOT / 'data' / 'processed' / 'meta.pkl'
if meta_path.exists():
    with meta_path.open('rb') as handle:
        meta = pickle.load(handle)
    vocab_size = meta['vocab_size']
    block_size = meta['block_size']
else:
    # Allows architecture experiments before preprocessing finishes.
    from poker_pipeline.tokenizer import PokerTokenizer
    vocab_size = len(PokerTokenizer().itos)
    block_size = 256

config = GPTConfig(vocab_size=vocab_size, block_size=block_size)
model = GPT(config)
print(config)
print(f'{model.get_num_params():,} non-position-embedding parameters')
```

## Poker-specific behavior already supported

- The forward pass accepts `loss_mask`, selecting only action and amount target tokens.
- `generate(..., allowed_token_ids=...)` can restrict sampling to legal action tokens.
- The defaults use a 256-token context and a small 6-layer model suitable for training from scratch.
